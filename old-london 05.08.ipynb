## Tokenization of the "description" column

```python
# step 1: text processing
tokenizer = CountVectorizer()
hashingTF = TfidfTransformer(use_idf=False)
idf = TfidfTransformer(use_idf=True)
k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
```

```python
# pipeline that processes text and then clusters
pipeline = Pipeline(steps=[
    ('tokenizer', tokenizer),
    ('hashingTF', hashingTF),
    ('idf', idf),
    ('kmeans', kmeans)
])
```

## LSTM baseline

```python
#https://machinelearningmastery.com/how-to-develop-lstm-models-for-time-series-forecasting/ 

def split_sequence(sequence, n_steps):
	X, y = list(), list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the sequence
		if end_ix > len(sequence)-1:
			break
		# gather input and output parts of the pattern
		seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
		X.append(seq_x)
		y.append(seq_y)
	return array(X), array(y)
```

```python
from numpy import array
X, y = split_sequence(df_train['temp'], 30)
```

```python
#number of steps will be the lagged features of 3
```

```python
n_steps = 30
n_features = 1
X = X.reshape((X.shape[0], X.shape[1], n_features)) #transforms the matrix from 3 values in each row to 3 values in each column
```

```python
def generate_predictions(model, initial_sequence, num_predictions):

    predictions = []
    current_sequence = np.array(initial_sequence)
    
    for _ in range(num_predictions):

        # this will predict the next 3 steps
        next_pred = model.predict(initial_sequence)

        # appens the predictions to the list
        predictions.append(next_pred)

        # updates the current_sequence with the new preds
        current_sequence = np.append(current_sequence, next_pred, axis=0)

    return np.array(predictions).reshape((-1, current_sequence.shape[-1]))
```
